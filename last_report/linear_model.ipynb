```python
import pandas as pd
from sklearn import linear_model
import numpy as np
```

```python
HEADER_ENGLISH = ('clock', 'core', 'RAM', 'ROM', 'Weight', 'screen', 'height', 'width', 'depth', 'wireless_charge', 'one_seg','ios', 'Android', 'value', 'out_camera', 'in_camera', 'battery', 'share')
```

```python
df = pd.read_csv('data2.csv', encoding='shift-jis'
                , names = HEADER_ENGLISH
                , index_col = 0
                , skiprows = 1
                )
```

```python
df
```

```python
clf = linear_model.LinearRegression()
```

```python
df_normalize = df.apply(lambda x:(x-np.mean(x))/(np.max(x)-np.min(x)))
```

```python
df_expect_share = df_normalize.drop('share', axis = 1)
df_expect_share = df_expect_share.drop('Android', axis = 1)
df_expect_share = df_expect_share.drop('core', axis = 1)
df_expect_share
```

```python
Y = df_normalize['share']
```

```python
Y
```

```python
clf.fit(df_expect_share, Y)
```

```python
df_expect_share
```

```python
Y
```

```python
print(pd.DataFrame({"Name":df_expect_share.columns
                   ,'Coefficients':clf.coef_ }))
```

```python
clf.coef_
```

```python
print(clf.intercept_)
```

```python
print(clf.score(df_expect_share, Y))
```
